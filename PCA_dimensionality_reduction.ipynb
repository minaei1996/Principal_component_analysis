```python
def zero_mean(x):
    """
    makes mean of each feature about to zero
    x = a pandas.dataframe with (observation, feature) format
    output = a numpy array with zero mean for each feature
    """
    data_array = np.array(x)
    feature_mean = np.mean(data_array, axis=0)
    zero_mean_data = data_array - feature_mean
    return zero_mean_data

def rescale(x):
    """
    nothing but a rescaling function to [0,1]
    x = a numpy array in (observation, feature) format
    output = rescaled x
    """
    feature_min = np.min(x, axis=0)
    feature_max = np.max(x, axis=0)
    rescaled_data = (x - feature_min) / (feature_max - feature_min)
    return rescaled_data

def error_z_pca(u,x,cut):
    """
    This function uses the u matrix from svd calculations to reduce the dimensions of data.
    u = [u, s, vh] from svd output.
    x = a numpy array in (feature, observation) format
    cut = u_reduced = u[:,:cut] part of u depending of cut
    output: z: reducted matrix of x.
            error: cost value of reduction. lower better.
    """
    u_reduced = u[:,:cut]
    z = np.dot(u_reduced.T, x)
    x_hat = np.dot(u_reduced,z)
    a = np.sum(np.sum((x_hat - x)**2, axis=1)/ x.shape[1])
    b = np.sum(np.sum(x**2, axis=1)/x.shape[1])
    error = a/b
    return z, error

def pca(x, threshold):
    """
    performs principal component analysis on a given input data.
    x = input data in (observation, feature) format
    threshold: maximun accepted error for dimensionality reduction
    output: z: reducted data
            error: cost of dimensionality reduction
            i: count of new features
    """
    zero_mean_x = zero_mean(x)
    rescaled_x = rescale(x)
    u, s, vh = np.linalg.svd(rescaled_x.T)
    for i in range(len(s)):
        z, error = error_z_pca(u, rescaled_x.T, i)
        if (error < threshold):
            return z.T, error, i
            break
```
